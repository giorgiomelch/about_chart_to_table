```python
from pathlib import Path
import json, shutil

ANNOTATION_PATH = Path("../../datasets/ChartQA/train/annotations")
IMGS_PATH = Path("../../datasets/ChartQA/train/png")

TRAININGSET_PATH = Path("../../training_data/ChartQA")
```

```python
for ann in ANNOTATION_PATH.iterdir():
    with open(ann, "r") as f:
        data = json.load(f)
        tipo = data.get("type")
        name = ann.with_suffix('.png').name
        img_path = IMGS_PATH / name
        shutil.copy2(img_path, TRAININGSET_PATH / tipo)
```
