```python
def copy_dir_content_to_dir(src_dir_path, target_dir_path, limit=None):
    count = 0
    target_dir_path.mkdir(parents=True, exist_ok=True) 
    for file in src_dir_path.iterdir():
        shutil.copy2(file, target_dir_path)
        if limit:
            if count>= limit:
                break
        count+=1
    print(f"Copiati {count} file.")
```

```python
def copia_immagini_casuali(percorso_sorgente, percorso_destinazione, quantita=2000):
    sorgente = Path(percorso_sorgente)
    destinazione = Path(percorso_destinazione)

    if not sorgente.exists() or not sorgente.is_dir():
        print(f"Errore: Il percorso {sorgente} non è una directory valida.")
        return

    tutti_i_file = list(sorgente.rglob('*.png'))

    # Controllo disponibilità
    numero_file = len(tutti_i_file)
    if numero_file < quantita:
        print(f"Errore: Trovate solo {numero_file} immagini. Ne servono {quantita}.")
        return

    # Selezione casuale
    selezione = random.sample(tutti_i_file, quantita)

    # Creazione cartella destinazione (compresi i genitori se mancano)
    destinazione.mkdir(parents=True, exist_ok=True)

    # Copia dei file
    for file_path in selezione:
        target_path = destinazione / file_path.name
        
        # Gestione collisioni nomi
        if target_path.exists():
            nuovo_nome = f"{file_path.stem}_{random.randint(1, 10000)}{file_path.suffix}"
            target_path = destinazione / nuovo_nome

        shutil.copy2(file_path, target_path)

    print(f"Completato: {quantita} immagini copiate in {destinazione}")
```
