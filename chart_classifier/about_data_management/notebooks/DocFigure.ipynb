```python
from pathlib import Path
import shutil

DOCFIGURE_IMAGE_PATH = Path("../../datasets/DocFigure/images")
DOCFIGURE_ANNOTATION_PATH = Path("../../datasets/DocFigure/annotation")
TRAINING_SET_PATH = Path("../../training_data/docfigure/")

ann_train_path = DOCFIGURE_ANNOTATION_PATH / "train.txt"
ann_test_path = DOCFIGURE_ANNOTATION_PATH / "test.txt"
```

```python
def get_filenames_for_label(annotation_path, findlabel):
    '''fornisce i filename delle immagini di una classe di grafici/immagini'''
    with open(annotation_path, "r") as f:
        img_names = []
        for line in f:
            filename, label = line.strip().split(",")
            label = label.lower()
            if findlabel in label:
                img_names.append(filename)
    print(f"Trovati {len(img_names)} file con label: {findlabel}")
    return img_names

def copy_file_to(img_names, from_dir, to_dir, label):
    from_dir = Path(from_dir)
    to_dir = Path(to_dir) / label
    to_dir.mkdir(parents=True, exist_ok=True)
    for imgname in img_names:
        from_path = from_dir / imgname
        to_path = to_dir / imgname
        shutil.copy2(from_path, to_path)
    print(f"Copiati {len(img_names)} file")
```

```python
docfigure_labels = ["medical", "natural", "mask", "3d",
                    "area", "bar", "box", "bubble", "contour",
                    "confusion", "heat map", "pareto", "pie", "polar",
                    "radar", "scatter", "surface", "vector", "venn",
                    "geographic map",
                    "table",
                    "flow", "block"]
```

```python
for label in docfigure_labels:
    img_names = get_filenames_for_label(ann_train_path, label)
    copy_file_to(img_names, DOCFIGURE_IMAGE_PATH, TRAINING_SET_PATH, label)
    img_names = get_filenames_for_label(ann_test_path, label)
    copy_file_to(img_names, DOCFIGURE_IMAGE_PATH, TRAINING_SET_PATH, label)
```

```python
for dir in TRAINING_SET_PATH.iterdir():
    label = dir.name
    file_count = len([f for f in dir.iterdir() if f.is_file()])
    print(f"{label}: {file_count}")
```
