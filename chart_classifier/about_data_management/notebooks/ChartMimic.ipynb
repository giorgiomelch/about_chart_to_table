```python
from pathlib import Path
import shutil
CHARTMIMIC_DATA = Path("../datasets/ChartMimic")
TRAINNG_DATA = Path("../training_data/ChartMimic")
```

```python
for dir in CHARTMIMIC_DATA.iterdir():
    if dir.is_dir():
        for img in dir.iterdir():
            filename = img.name
            label = filename.split("_")[0]
            target_dir = TRAINNG_DATA / label 
            target_file = TRAINNG_DATA / label /filename
            Path.mkdir(target_dir, parents=True, exist_ok=True)
            print(target_file)
            shutil.copy2(img, target_file)
```
