```python
from pathlib import Path
import shutil
import random

PMC_OA_DIR_PATH = Path("../../datasets/PMC-OA_dataset/caption_T060_filtered_top4_sep_v0_subfigures/")
```

```python
images = list(PMC_OA_DIR_PATH.glob("*.jpg"))
random.shuffle(images)

keep = 20_000
to_delete = images[keep:]

print(f"Eliminazione di {len(to_delete)} file in corso...")

for image in to_delete:
    image.unlink()
print("Pulizia completata: La cartella contiene: ", len(list(PMC_OA_DIR_PATH.iterdir())))
```

```python
import pandas as pd
df = pd.read_csv("../../datasets/PMC-OA_dataset/caption_T060_filtered_top4_sep_v0_subfigures.csv")
df = df[df["prediction"]!="image"]
df.size
df.head
```

```python
from PIL import Image
TARGET_PATH = "../../../data_management/training_data/all/medical_image"
for filename in df["filename"]:
    src_path = PMC_OA_DIR_PATH / filename
    shutil.copy2(src_path, TARGET_PATH)
```
