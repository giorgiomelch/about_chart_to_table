```python
from pathlib import Path
import shutil
import random

MPP_DIR_PATH = Path("../../datasets/MPP")

MOLECULE_DIR_PATH = Path("../../training_data/all/molecule")
if not MOLECULE_DIR_PATH.exists():
    MOLECULE_DIR_PATH.mkdir(parents=True, exist_ok=True)
```

```python
max_images = 4000
images_paths = list(MPP_DIR_PATH.rglob("*.png"))
random.shuffle(images_paths)
for img_path in images_paths:
    shutil.copy2(img_path, MOLECULE_DIR_PATH)
    max_images = max_images - 1
    if max_images==0: break
```
